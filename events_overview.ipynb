# **SLOT EVENTS DIAGRAMS**

```python
from IPython.display import Image, display

from diagrams import Diagram
from diagrams.c4 import Container, Relationship
```

## **USER SERVICE EVENTS OVERVIEW**

```python
graph_attr = {
    "splines": "spline",
}

with Diagram("User Service Events Overview", direction="LR", graph_attr=graph_attr, show=False):
    user_service = Container(
        name="User Service",
        technology="Spring Boot",
        description="Manages users and tenants",
    )
    scheduling_service = Container(
        name="Scheduling Service",
        technology="Spring Boot",
        description="Manages working hours, slots and bookings",
    )
    search_service = Container(
        name="Search Service",
        technology="Spring Boot",
        description="Manages denormalized data for search",
    )

    user_service >> Relationship("ProfessionalCreated", color="#fd5847") >> scheduling_service # projection pattern
    user_service >> Relationship("ProfessionalDeleted", color="#fd5847") >> scheduling_service # projection pattern
    user_service >> Relationship("TenantCreated", color="#fd5847") >> [scheduling_service, search_service]
    user_service >> Relationship("TenantUpdated", color="#fd5847") >> search_service
    user_service >> Relationship("TenantDeleted", color="#fd5847") >> [scheduling_service, search_service]

display(Image("user_service_events_overview.png"))
```
